```python
import chromedriver_binary
from webdriver_manager.chrome import ChromeDriverManager

import selenium
from selenium import webdriver as wd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import random
import time
import datetime
```

```python
url = 'https://www.newegg.ca/p/0TH-04DJ-00006?Item=9SIAD1N8V54134&Description=headphones&cm_re=headphones-_-9SIAD1N8V54134-_-Product'

first_name = 'Shrey'
last_name = 'Verma'
address_first_line = '50 Lena Drive'
city = 'RICHMOND HILL'
province = 'ON'
postal_code = 'L4S 2V1'
phone = '9057808166'
email = 'shreyverma@hotmail.com'
cardholder_name = 'Jack Bui'
cardnumber = '1234567891234567'
expiry_month = 6 
expiry_year = 2022
cv2 = '123'
```

```python
wd = wd.Chrome(ChromeDriverManager().install())
wd.implicitly_wait(10)
```

```python
def NewEgg(url,first_name,last_name, address_first_line , city , province , postal_code , phone , email 
           ,cardholder_name , cardnumber , expiry_month , expiry_year , cv2):
    
    year = expiry_year - datetime.datetime.now().year
    
    wd.get(url)
    
    # Add to Cart and Checkout 
    random_wait_time = random.randrange(1.0, 3.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    add_to_cart_button = wd.find_element_by_xpath('//*[@id="ProductBuy"]/div/div[2]/button')
    add_to_cart_button.click()
    
    #Might need to click no_thanks if it pops up
    try: 
        random_wait_time = random.randrange(1.0, 4.0)
        time.sleep(random_wait_time)
        print(random_wait_time)
        
        no_thanks = wd.find_element_by_xpath('//*[@id="modal-intermediary"]/div/div/div/div[3]/button[1]')
        no_thanks.click()
        view_cart_and_checkout = wd.find_element_by_xpath('//*[@id="modal-intermediary"]/div/div/div[2]/div[2]/button[2]')
        view_cart_and_checkout.click()
        
    except:
        random_wait_time = random.randrange(1.0, 4.0)
        time.sleep(random_wait_time)
        print(random_wait_time)
        
        view_cart_and_checkout = wd.find_element_by_xpath('//*[@id="modal-intermediary"]/div/div/div[2]/div[2]/button[2]')
        view_cart_and_checkout.click()
    
    # Checkout 
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)

    secure_checkout = wd.find_element_by_xpath('//*[@id="app"]/div[1]/section/div/div/form/div[2]/div[3]/div/div/div[3]/div/button')
    secure_checkout.click()
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    guest_checkout = wd.find_element_by_xpath('//*[@id="app"]/div/div[2]/div[2]/div/div/div[1]/form[2]/div[2]/div/button')
    guest_checkout.click()
           
    # Shipping Info
    random_wait_time = random.randrange(2.0, 5.0)
    time.sleep(random_wait_time)
    print(random_wait_time) 
    try:
        add_new_address = wd.find_element_by_xpath('//*[@id="shippingItemCell"]/div/div[2]/div[2]/div/div[2]/div[2]/div/div/span')
        add_new_address.click()
    except:
        add_new_address = wd.find_element_by_xpath('//*[@id="shippingItemCell"]/div/div[2]/div[2]/div/div[2]/div[2]/div/div/i')
        add_new_address.click()
    
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    first_name_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[1]/input')
    first_name_section.send_keys(first_name)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    last_name_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[2]/input')
    last_name_section.send_keys(last_name)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    address_first_line_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[5]/input')
    address_first_line_section.send_keys(address_first_line)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    city_section =  wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[7]/input')
    city_section.send_keys(city)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    province_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[8]/label[2]/select')
    Select(province_section).select_by_value(province)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    postal_code_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[9]/input')
    postal_code_section.send_keys(postal_code)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    phone_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[11]/input')
    phone_section.send_keys(phone)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    email_section = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[2]/form/div[2]/div[14]/input')
    email_section.send_keys(email)
    
    
    save = wd.find_element_by_xpath('//*[@id="app"]/div/div/div/div/div[3]/button[2]')
    save.click()
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)

    try: 
        # May need to click
        use_address = wd.find_element_by_xpath('/html/body/div[7]/div/div/div/div/div[3]/button[2]')
        use_address.click()
        continue_to_delivery = wd.find_element_by_xpath('//*[@id="shippingItemCell"]/div/div[3]/button')
        continue_to_delivery.click()
    except:
        continue_to_delivery = wd.find_element_by_xpath('//*[@id="shippingItemCell"]/div/div[3]/button')
        continue_to_delivery.click()
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    continue_to_payment= wd.find_element_by_xpath('//*[@id="deliveryItemCell"]/div/div[3]/button')
    continue_to_payment.click()
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    add_credit_card = wd.find_element_by_xpath('//*[@id="paymentItemCell"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div')
    add_credit_card.click()
    
    # Credit Card 
    WebDriverWait(wd, 20).until(EC.frame_to_be_available_and_switch_to_it((By.XPATH,'/html/body/div[7]/div/div/div/div/div/iframe')))
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    WebDriverWait(wd, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[6]/div/div[2]/div[1]/div[1]/input'))).send_keys(cardholder_name)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    number_section = wd.find_element_by_xpath('/html/body/div[6]/div/div[2]/div[1]/div[2]/input')
    
    number_section.send_keys(Keys.BACKSPACE * 3,  cardnumber)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    month_section = wd.find_element_by_xpath('//*[@id="app"]/div/div[2]/div[1]/div[4]/label[2]/select')
    Select(month_section).select_by_index(expiry_month)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    year_section = wd.find_element_by_xpath('//*[@id="app"]/div/div[2]/div[1]/div[5]/label/select')
    Select(year_section).select_by_index(year)
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    save_credit = wd.find_element_by_xpath('//*[@id="app"]/div/div[3]/button[2]')
    save_credit.click()
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)

    cv2 = wd.find_element_by_xpath('//*[@id="paymentItemCell"]/div/div[2]/div[2]/div[2]/div[2]/div[2]/div/div/label/div[4]/input')
    cv2.send_keys('123')
    
    random_wait_time = random.randrange(1.0, 4.0)
    time.sleep(random_wait_time)
    print(random_wait_time)
    
    review_order = wd.find_element_by_xpath('//*[@id="paymentItemCell"]/div/div[3]/button')
    review_order.click()
    
    
    return True 
```

```python
NewEgg(url ,first_name,last_name, address_first_line , city , province , postal_code , phone , email 
           ,cardholder_name , cardnumber , expiry_month , expiry_year , cv2)
```
